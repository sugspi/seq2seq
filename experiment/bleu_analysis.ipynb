```python
#bleu evaluation
len_inp = 1500#len(test_input_data)
sum_score = 0
complete = 0
comp_texts = []
one = 0
two = 0
three = 0
four = 0
five = 0
six = 0
seven = 0

ones = []
twos = []
threes = []
fours = []
fives = []
sixs = []
sevens = []

for seq_index in range(len_inp-1):
    fname = 'char/c2l/result'+str(seq_index)+'.txt'
    f = open(fname)
    lines = f.readlines()
    f.close()

    orign_text = lines[0].rstrip()
    if(orign_text[-1]!='.'):
        orign_text = orign_text + '.'
    decoded_sentence = lines[1].rstrip()
    score = sentence_bleu([orign_text],decoded_sentence)
    sum_score += score
    
    if(score <= 0.2):
        one += 1
        ones.append([orign_text,decoded_sentence])
    elif (score <= 0.3):          
        two += 1
        twos.append([orign_text,decoded_sentence])
    elif (score <= 0.4):          
        three += 1
        threes.append([orign_text,decoded_sentence])
    elif(score <= 0.5):
        four += 1
        fours.append([orign_text,decoded_sentence])
    elif (score <= 0.6):          
        five += 1
        fives.append([orign_text,decoded_sentence])
    elif (score <= 0.7):          
        six += 1
        sixs.append([orign_text,decoded_sentence])
    else:
        seven += 1
        sevens.append([orign_text,decoded_sentence])
        
    
    if(orign_text == decoded_sentence):
        complete += 1
        comp_texts.append(orign_text)
    
    print(orign_text)
    print(decoded_sentence)
    print(score,"\n")
    
    
    #f.write(output_texts[seq_index])
    #f.write(decoded_sentence.strip()+'\n')
    #f.close()
    #print('Input sentence:', input_texts[seq_index])
    #if (seq_index%100) == 0 :
    #    print('Decoded sentence:', decoded_sentence)
    #    print('Answer sentence:', output_texts[seq_index])
print('bleu score: ',(sum_score/len_inp))
print('complete: ',complete)
print(comp_texts)
```
