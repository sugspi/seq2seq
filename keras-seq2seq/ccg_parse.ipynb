```python
import bs4
import glob
import os
import sys
import concurrent.futures
import pickle
import MeCab
import json

```

```python
def getKana(chasen):
  es = chasen.split("\n")
  es.pop()
  es.pop()
  return "".join( map(lambda x:x.split("\t")[1], es) )
```

```python
#dict_pklを作るための関数
def gen_courpus():
    
  f = open('/Users/guru/MyResearch/sg/kyotoU/json/ku_inp_1019.json')
  jsonData = json.load(f)

  mcb = MeCab.Tagger("-Owakati")
  sens1 = [mcb.parse((jsonData[i]['formula'])).strip() for i in jsonData]
  sens2 = [mcb.parse(((jsonData[i])['text']).replace('\n', '')).strip() for i in jsonData]
  
  m = MeCab.Tagger("-Ochasen")
  with open("ku_corpus.txt", "w") as f:
    for i in range(len(sens1)):
      head = getKana( m.parse( sens1[i] ) )
      tail = getKana( m.parse( sens2[i] ) )
      i = i+1;

      if len(tail) <= 50 and len(tail) <= 50:
        j = "___SP___".join([head, tail])
        f.write(j + "\n")
        print(j)
```

```python
gen_courpus()
```

```python
""" 128位以下は削る """
def distinct(depth=128, target="ku_corpus.txt", output="ku_corpus.distinct.txt"):
  c_f = {}
  for c in open(target, "r").read() :
    if c_f.get(c) is None:
      c_f[c] = 0
    c_f[c] += 1
  for e, (c, f) in enumerate(sorted( c_f.items(), key=lambda x:x[1]*-1)):
    print(e, c, f)
  c_i = {c:e for e, (c,f) in  enumerate(sorted( c_f.items(), key=lambda x:x[1]*-1)[:depth]) }
  #print(c_i)
  open("c_i.pkl", "wb").write( pickle.dumps( c_i ) )
  with open(target, "r") as r, open(output, "w") as w:
    for ln in r:
      ln = ln.strip()
      distict = "".join( filter(lambda x: x in c_i, list(ln) ) )
      #w.write( distict + "\n" )
      print(distinct)
```
